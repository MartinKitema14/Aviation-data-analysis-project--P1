DESCRIPTIVE AND EXPLORATORY ANALYSIS

```python
#Creating a dataframe that distributes the total accident numbers as per each make of thge air crafts. The company should avoid the make with more accident history.

Make= new_df['Make'].value_counts().to_frame()
```

```python
Make.head()
```

```python
import matplotlib.pyplot as plt

# Plot the 'Make' column value counts as a bar chart
ax =new_df['Make'].value_counts() \
.head() \
.plot(kind='bar', title="Make with Most Accidents")

ax.set_xlabel('Make') 
ax.set_ylabel('Accidents count')

# Show the plot
plt.show()
```

```python
#Creating a dataframe that shows if engines builty by ameteuir had impact on the number of accidents.

Amateur_Built= new_df['Amateur.Built'].value_counts().to_frame()
```

```python
Amateur_Built.head()
```

```python
import matplotlib.pyplot as plt

# Plot the 'Amateur_Built' column value counts as a bar chart
ax =new_df['Amateur.Built'].value_counts() \
.head() \
.plot(kind='bar', title="Amateur_Built")

ax.set_xlabel('Amateur_Built') 
ax.set_ylabel('Accidents count')

# Show the plot
plt.show()
```

```python
#Creating dataframe to determine the engine type with most accidents
Engine_Type= new_df['Engine.Type'].value_counts().to_frame()
Engine_Type.head()
```

```python
import matplotlib.pyplot as plt

# Plot the 'Make' column value counts as a bar chart
ax =new_df['Engine.Type'].value_counts() \
.head() \
.plot(kind='bar', title="Engine Types with Most Accidents")

ax.set_xlabel('Engine_Type') 
ax.set_ylabel('Accidents count')

# Show the plot
plt.show()
```

```python
#Distribution of the accients in different countries
Country= new_df['Country'].value_counts().to_frame()
Country.head()
```

```python
#Aircraft damage
Aircraft_damage= new_df['Aircraft.damage'].value_counts().to_frame()
Aircraft_damage.head()
```

```python
# Mean of Total fatal injuries
Total_fatal_injuries = pd.DataFrame({'Mean_Total_Fatal_Injuries': [new_df['Total.Fatal.Injuries'].mean()]})
print(Total_fatal_injuries.head())
```
